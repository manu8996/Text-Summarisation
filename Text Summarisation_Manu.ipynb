```python

    
def createbody(sections,body_text):    
    if sections==[]:
            return body_text
    else:
        for section in sections:
            if (section.text!="  References " and section.text!="  Conclusion "):
                body_text=createbody(section.getchildren(),body_text)
                if (section.text!=None and section.tag!='HEADER'):
                    body_text +=section.text 
            
    return body_text    
    
body=root.find('BODY')
body_text = ''
sections=body.getchildren()
body_text=createbody(sections,body_text)
print(body_text)

text_file = open("cmplg-xml/9404004.txt", "w")

text_file.write(body_text)

text_file.close()    
        
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk_lemmatized = []
word_net_lemmatizer = WordNetLemmatizer()
for line in nltk_token_lower:
    tokens = word_tokenize(line)
    lines = nltk.pos_tag(tokens)
    lemma = ""
    #print(len(lines))
    for i in range(len(lines)):
        #print(lines[i][0],lines[i][1])
        try:
            lemma += " " +word_net_lemmatizer.lemmatize(lines[i][0],get_wordnet_pos(lines[i][1]))
            #print(word_net_lemmatizer.lemmatize(lines[i][0],get_wordnet_pos(lines[i][1])))
        except:
            lemma += " " + lines[i][0]
    nltk_lemmatized.append(lemma)
    print(lemma)
```

```python
#import re
#def replaceMultiSpace(text):
 #   text = re.sub(r"(\\n)\1+", ' ', text)
  #  return text
#nltk_tokens = replaceMultiSpace(nltk_tokens)

j =1
nltk_processed = []
for line in nltk_lemmatized:
    #prinat(line)
    print("\n\nLine No. ",j)
    k = line.split('\n')
    lines = ''
    for i in range(len(k)):
        if k[i] != '':
            lines += " " 
            lines += k[i]
    print(lines)
    nltk_processed.append(lines)
    j = j+1
    
```
